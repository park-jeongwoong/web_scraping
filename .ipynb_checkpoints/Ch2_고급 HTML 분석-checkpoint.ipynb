```python
# 2.2. find() & findAll() 함수로 필터 적용하여 스크래핑하기

from urllib.request import urlopen
from bs4 import BeautifulSoup
import ssl

context = ssl._create_unverified_context()
html = urlopen('https://pythonscraping.com/pages/warandpeace.html', context = context)
bs = BeautifulSoup(html, 'html.parser')

# findAll() 함수의 사용. 이것을 통해 아래 태그로 둘러싸인 텍스트들만 추출할 것이다.

"""<span class="green"></span>"""

# 이러한 태그는 css로 html을 저렇게 둘러싸 green 색깔로 글씨를 표현할 수 있게 한다.


# 그렇다면, 이것들을 쭉 한번 출력해보자.
# 클래스 객체인 bs에 속한 findAll() 함수는 bs.findAll(tagName, tagAttributes)와 같은 형식으로 호출한다.
# 이전까지는 bs.tagName 으로 처음 나타난 태그를 추출했다면, 
# 이번에는 모두 추출하며 그 태그의 Attributes까지 지정해 준 것이다.

nameList = bs.findAll('span',{'class':'green'}) 
for name in nameList: # 해당 클래스 함수를 이용해 추출한 것은 '태그 객체'(중요)이고 리스트 형태이다.
    print(name.get_text()) # get_text() 는 태그를 제외하고 콘텐츠만 출력할 때 사용할 수 있다.
    # .get_text()는 현재 문서에서 모든 태그를 제거하고 유니코드 텍스트만 들어 있는 문자열을 반환한다.
    # 유니코드는 컴퓨터 산업에서 표준적으로 정의하고 있는 문자의 모음이라고 보면 된다.
    # 항상 블록 형태 일때보다는 마지막 BeautifulSoup 클래스의 객체에 사용하는 게 유리하다.
    # 최종 데이터를 출력하거나 저장, 조작하기 '직전'에만 써야한다고 생각하자.
    
"""
저 사이트 css는, 등장인물의 이름에만 초록색 칠을 해놓았다.
따라서 이와 같이 findAll 함수를 통해 초록색 Attribute만 모아놓은 것은 등장인물의 이름만 추출했다고 보면 된다.
"""
```

```python
# 2.2.1. find()와 findAll()의 매개변수로는 무엇이 있을까?

"""
find()와 findAll()은 HTML 페이지에서 원하는 태그를 지정한 속성에 따라 필터링할 수 있다.
매개변수로는 tag와 attributes을 주로 사용한다.

1) tag 매개변수_
:태그 이름인 '문자열'을 넘기거나, 태그 이름으로 이루어진 '리스트'를 넘길 수도 있다.

bs.findAll({'h1', 'h2', 'h3', 'h4', 'h5', 'h6'}) 
# 대괄호이기 때문에 '집합(set)'인 것 같지만, 사실 리스트가 맞고 모든 매개변수에 괄호를 이렇게 치는 것 같다.

2) attributes 매개변수_
:속성으로 이루어진 파이썬 '딕셔너리'를 받고, 그 중 하나에 일치하는 태그를 찾아 준다.(or 개념)

bs.findAll('span',{'class':{'green','red'}})

3) text 매개변수_
: 태그에 둘러싸인 것들 중, text와 정확히 일치하는 단어를 추출해준다. 다음과 같이 작성할 수 있다.

nameList = bs.findAll(text = 'the prince')
print(len(nameList))

4) limit 매개변수_
: find()와 findAll()을 구분하는 매개변수로, find 는 이 값이 1이라고 보면 된다.
위에서부터 그 태그를 몇 개나 찾을 것인지 묻는 매개변수이다.
                      
5) keyword 매개변수_
: attributes 매개변수와 본질적으로 동일하다. 어떤 속성이 포함된 태그를 추출할 것인지 묻는 매개변수이다.
차이점이라면, attributes 매개변수는 or 필터로 동작하지만, keyword 매개변수는 and 필터처럼 동작한다는 점이다.

nameList = bs.findAll('span', id = 'title', class_ = 'text) # span은 태그 지정, 그 뒤의 두개는 모두 keyword.
# 참고로 class 뒤에 밑줄을 그어준 이유는, class라는 키워드 자체가 파이썬의 예약어이기 때문이다.(list 같은..)

nameList = bs.findAll('span',{'id' : 'title', 'class' : 'text'}) 이는 같은 의미의 nameList를 
Attributes 매개변수를 이용한 결과이다.
"""
```

```python
# 2.2.3. 트리 이동

"""
#.
findAll()이라는 함수는 뷰티풀솦 클래스 객체 내의 함수로서, 이름과 속성에 따라 태그를 찾아냈다.

<span class="green"></span>이라는 태그가 있다고 할 때,
태그의 이름은 'span'(tag 매개변수), 속성은 'class' : 'green'(attributes 매개변수) 였다.

#.
그렇다면 위치에 따라 태그를 찾으려면 어떻게 해야할까?
'트리 내비게이션'을 통해 그것을 수행할 것인데, 우리는 사실 이미 유사한 코드를 작성한 적이 있다.
첫 번째 챕터 주피터 노트북의 3번째 쉘에 가면 아래와 유사한 형태의 코드를 찾아 볼 수 있을 것이다.

bs.tag.subTag.anotherSubTag

트리를 단방향으로 이동하는 것인데, 태그를 하나 찾고, 그 아래 태그, 그 아래 태그로 이동하는 것이다.
다시 개념의 위치감각을 살려보자면, 뷰핕풀솦 클래스 객체 내의 하나의 기능(?)으로서 태그를 이동하는 것이다.

우선 아직 html의 구조가 확실히 익지 않았기 때문에 직접 파싱한 객체를 print 해보면서,

nameList = bs.findAll('span')과 어떤 차이가 있었는지를 느껴보자.

"""
# print(bs)
"""
자. urlopen으로 사이트에서 html 객체를 만들고 뷰티풀솦 클래스에서 구문 분석기로 'html.parser'를 이용하여,
url의 html 파일만 bs 객체에 담아놓았다. 윗 코드는 단순히 그것을 print한 것에 불과하다.

여기서 findAll('span') 클래스함수는 단순히 tag 매개변수로 'span'을 받고
span 태그를 몽땅 찾아준 것이다. 그 이후에 attributes에 해당하는 것을 or 필터로 건져냈다.

그것에 비한다면, 이번 '트리 이동'은 그것보다 조금더 구조적인 차원으로 접근한다.
아래 출력 결과를 보면 html 파일은 상당히 구조적이다.
html 태그 안에 head, body 등의 큰 태그가 있고, 그 안에 style이라던지 작은 태그로 하나씩 세분화된다.

지금부터 이러한 구조에서 '트리 이동'을 하는 방법을 알아보려는 것이다.

"""

# 자. 이제 책에 따라 https://pythonscraping.com/pages/page3.html의 온라인 쇼핑 사이트의 html 파일로
# 트리를 이동하는 방법을 실습해보자.
```

```python
#. 자식과 자손의 구분

"""
자식은 바로 아래, 자손은 그 아래 전부를 포함하는 개념이다.

당연히 일반적으로 뷰티풀소프 함수는 항상 현재 선택된 태그의 자손을 다룬다.
예를들어 bs.body.h1은 자손인 h1 태그를 선택하지, body 바깥에 있는 태그에 대해서 동작하지 않는다는 뜻이다.

트리 이동은 findAll()과도 같이 쓰일 수 있는데,
bs.div.findAll("img")는 문서의 첫 번째 div 태그를 찾고, 그 태그의 자손인 img 태그의 목록을 findAll()해준다.

자식만 찾을 때에는 .children을 사용한다. (반대로 자손은 descendants로 통용된다.)

"""
from urllib.request import urlopen
from bs4 import BeautifulSoup
import ssl

context = ssl._create_unverified_context()
html = urlopen('https://pythonscraping.com/pages/page3.html', context = context)
bs = BeautifulSoup(html, 'html.parser')

for child in bs.find('table',{'id' : 'giftList'}).children: 
    print(child)
    
# 이 때 만약 children() 대신 descendants() 함수를 썼다면 테이블에 포함된 태그가 대폭 늘어났을 것이다!
# 직접 바꿔서 쳐보면 느낄 수 있다!


# 이외에도 형제 등을 이용하여 트리를 이동할 수 있다.

# bs.find('table', {'id' : 'giftList'}).tr.next_sibling

# 위 코드는 bs 객체에서 table 이라는 태그 중 attribute이 id : giftLIst인 태그를 찾아낸 뒤,
# 그 중 tr이라는 자손 코드에 들어간 이후 nextsibling 을 이용해 트리를 이동한 것이다.(동등한 높이의 다음 태그)
# next_sibling외에도 previous_sibling가 있고 두 형제코드 모두 뒤에 s를 붙여 이터레이터로 만들 수도 있다.

# 아래로 가는 것 이외에도 children의 반대격으로 .parent와 .parents도 있다.
```

```python
# 2.4. 정규 표현식과 BeautifulSoup

# 문자열 매개변수를 받는 대부분의 함수는 정규표현식도 매개변수로 받을 수 있는데, 
# 실제 웹스크래핑에 한번 적용해보자.

import re # 정규표현식을 사용할 수 있도록 import

html = urlopen('http://www.pythonscraping.com/pages/page3.html', context = context)
bs = BeautifulSoup(html, 'html.parser')
images = bs.findAll('img',{'src':re.compile('\.\.\/img\/gifts\/img.*\.jpg')})
for image in images:
    print(image['src']) # 추출한 태그들의 list를 images에 담았고 하나씩 추출하고 있다.
                        # image 태그 중에서 src 인덱스를 갖고 있는 것을 그대로 뽑아준 것이다.
                        # .get_text는 태그 안의 콘텐츠를 뽑아오는 것이지만, 이것은 태그의 속성이므로
                        # 위 함수가 아니라, 인덱싱을 사용해준 모습이다.
    
# 정규 표현식은 표현식 어디든 매개변수로 삽입할 수 있기 때문에, 매우 유연하게 원하는 요소를 찾을 수 있게 한다.
```

```python
# 2.5. 속성에 접근하기

"""
지금까지는 태그에 접근하고 필터링하는 법, 그 안에 들어 있는 콘텐츠에 접근하는 법을 알아보았다.
하지만, 그 태그의 콘텐츠가 아니라 그 속성(attribute)에 관심이 있을 때가 자주 있다.
위의 image['src']도 'img'태그의 콘텐츠가 아니라, 그 태그의 'src'속성을 추출한 것이다.

이때, 아래 함수를 통해 태그 객체에서 속성 목록에 접근할 수 있다.

myTag.attrs['myAttribute']

바로 위 쉘에서 image['src']를 image.attrs['src']로 치환해주어도 결과가 같은 것을 확인할 수 있다.

"""
```

```python
# 2.6. 람다 표현식

"""
람다 함수의 문법을 잘 이해하고 적절히 태그 객체를 불리언(T/F)로 전달할 수 있다면,
함수의 find() 함수의 매개변수를 기억하지 못해도 원하는 일을 할 수 있다.

예를 들면 다음과 같은 것이다.

다음 코드는 속성이 정학히 두 개인 태그를 모두 가져오는 코드이다.
"""

Tags = bs.findAll(lambda tag: len(tag.attrs) == 2)
for Tag in Tags:
    print(Tag.get_text())
    
# 이 때, 람다함수의 매개변수는 반드시 태그 객체만을 넣고, 불리언만 반환할 수 있다는 제약이 있다.
# 여기서는 그 제약에 따라, 속성의 개수가 2개인 것들을 Tags 리스트에 담고,
# 하나씩 뽑아 그 태그의 콘텐츠를 출력한 것이다.(이때, 콘텐츠가 없는 것들은 빈칸만 출력되었을 것이다.)
# 실제로 'img' 태그가 속성값만 있고, 콘텐츠는 없었던 대표적인 태그이다. 
```
